#Differences Between performance from year to year
###Add Differences to Table (Girls)

```python
diff1_3 = []
diff2_3 = []
diff1_2 = []
for results in only_three['list_results']:
    first = results[0]
    second = results[1]
    third = results[2]
    
    diff1_3.append(third - first)
    diff1_2.append(second - first)
    diff2_3.append(third - second)

imprv1_3 = [1 if d > 0 else 0 for d in diff1_3]
imprv2_3 = [1 if d > 0 else 0 for d in diff2_3]
imprv1_2 = [1 if d > 0 else 0 for d in diff1_2] 

only_three['diff1_3'] = diff1_3
only_three['diff2_3'] = diff2_3
only_three['diff1_2'] = diff1_2
only_three['imprv1_3'] = imprv1_3
only_three['imprv2_3'] = imprv2_3
only_three['imprv1_2'] = imprv1_2
len(only_three.index)
```

```python
bdiff1_3 = []
bdiff2_3 = []
bdiff1_2 = []
for results in boys_three['list_results']:
    first = results[0]
    second = results[1]
    third = results[2]
    
    bdiff1_3.append(third - first)
    bdiff1_2.append(second - first)
    bdiff2_3.append(third - second)

bimprv1_3 = [1 if d > 0 else 0 for d in bdiff1_3]
bimprv2_3 = [1 if d > 0 else 0 for d in bdiff2_3]
bimprv1_2 = [1 if d > 0 else 0 for d in bdiff1_2] 

boys_three['diff1_3'] = bdiff1_3
boys_three['diff2_3'] = bdiff2_3
boys_three['diff1_2'] = bdiff1_2
boys_three['imprv1_3'] = bimprv1_3
boys_three['imprv2_3'] = bimprv2_3
boys_three['imprv1_2'] = bimprv1_2
len(boys_three.index)
```

#Make Section and Class Numerical Classifiers
###Classifications based on Section, Top Teams, and Class (Girls)

```python
master_girls = master_girls_2013.copy()
master_girls['top_schools'] = [1 if school in top_teams else 0 for school in master_girls_2013.School]

sections = []
for section in master_girls.Section:
    #print section
    if section[0] == '0':
        sections.append(section[1])
    elif section[0] == '1':
        sections.append(section[0:2])
    else:
        if section == 'AIS':
            sections.append(12)
        elif section == 'CHS':
            sections.append(13)
        else:
            sections.append(14)

classes = []
for cl in master_girls.Class:
    if cl == 'A':
        classes.append(1)
    elif cl == 'B':
        classes.append(2)
    elif cl == 'C':
        classes.append(3)
    elif cl == 'D':
        classes.append(4)
    elif cl == 'AIS':
        classes.append(5)
    elif cl == 'CHS':
        classes.append(6)
    else:
        classes.append(7)

        
master_girls['Section'] = sections
master_girls['Class'] = classes
master_girls.head()
```

#Run Random Forest Classifer
###Feature Data and Target Data

```python
##Boys
X_girls = master_girls[['1_Rating', 'diff1_2', 'Class', 'Section', 'top_schools', '1_Place', '2_Place', '3_Place']]
Y_girls = master_girls['imprv1_3']

X_boys = master_boys[['1_Rating', 'diff1_2', 'Class', 'Section', 'top_schools', '1_Place', '2_Place', '3_Place']]
Y_boys = master_boys['imprv1_3']
```

##Cross Validation Scores (Girls)

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import cross_val_score

gscores_trees = []

for i in np.arange(1,41,1):
    forest = sklearn.ensemble.RandomForestClassifier(n_estimators=i)
    gscores = sklearn.cross_validation.cross_val_score(forest, X_girls,  Y_girls, cv=10)
    gscores_trees.append(gscores)
```

```python
sns.set()
sns.set_context("notebook")
plt.figure(figsize=(12, 6))


sns.boxplot(gscores_trees, np.arange(1,41,1))
sns.set_context('poster')
plt.title("Cross-Validation Score vs. # of Trees for Random Forest Classifier")
plt.ylabel("Cross-Validation Score")
plt.xlabel("# of Trees")
plt.show()
```

##Cross Validation Score (Boys)

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import cross_val_score

bscores_trees = []

for i in np.arange(1,41,1):
    forest = sklearn.ensemble.RandomForestClassifier(n_estimators=i)
    bscores = sklearn.cross_validation.cross_val_score(forest, X_boys,  Y_boys, cv=10)
    bscores_trees.append(bscores)
```

```python
sns.set()
sns.set_context("notebook")
plt.figure(figsize=(12, 6))


sns.boxplot(bscores_trees, np.arange(1,41,1))
sns.set_context('poster')
plt.title("Cross-Validation Score vs. # of Trees for Random Forest Classifier")
plt.ylabel("Cross-Validation Score")
plt.xlabel("# of Trees")
plt.show()
```

#Plot Decision Surface of Random Forest Classifier and Random Decision Tree
###Plots (Girls)

```python
from sklearn.tree import DecisionTreeClassifier
import sklearn.linear_model
import sklearn.svm

def plot_decision_surface(clf, X_train, Y_train):
    plot_step=0.1
    
    if X_train.shape[1] != 2:
        raise ValueError("X_train should have exactly 2 columnns!")
    
    x_min, x_max = X_train[:, 0].min() - plot_step, X_train[:, 0].max() + plot_step
    y_min, y_max = X_train[:, 1].min() - plot_step, X_train[:, 1].max() + plot_step
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))

    clf.fit(X_train,Y_train)
    if hasattr(clf, 'predict_proba'):
        Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:,1]
    else:
        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])    
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=plt.cm.Reds)
    plt.scatter(X_train[:,0],X_train[:,1],c=Y_train,cmap=plt.cm.Paired)
    plt.xlabel(X_girls_important.columns[0])
    plt.ylabel(X_girls_important.columns[1])
    plt.show()
    
X_girls_important = master_girls[['1_Rating', 'diff1_2']]
forest = sklearn.ensemble.RandomForestClassifier(n_estimators=15)
forest.fit(X_girls_important, Y_girls)
decision_tree = sklearn.tree.DecisionTreeClassifier()
decision_tree.fit(X_girls_important, Y_girls)

girls_important_matrix = X_girls_important.as_matrix()
print "Random Forest Classifier: "
plot_decision_surface(forest, girls_important_matrix, Y_girls)
print ""
print "Decision Tree Classifier: "
plot_decision_surface(decision_tree, girls_important_matrix, Y_girls)
```

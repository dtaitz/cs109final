# Trends in Performance over High School Careers in NYS XC

## Background and Motivation

<p>In high school, 2 of our 3 project members participated in track and cross country and the third being a fan of sports analysis. As running is a largely mental sport, it can be very consuming for the individual and thus many athletes try to find trends to help them improve as a runner or predict results from year to year. Luckily, since track and cross country are timed events, data is easily observed and also maintained by a host of different websites. In New York State, where David participated in high school cross country, tullyrunners.com, ny.milesplit.com, and dyestat.com hold a large database of results. In the case of tullyrunners.com, the founder of the website also conveniently calculates "speed ratings" on cross country results to normalize for the various variables (course difficulty, level of competition, weather, etc) that can affect one's time on a cross country 5K course. His modeling of cross country has been widely accepted in the track and field community and thus a reliable measure to run analysis on.
<br>
<p> One particular trend that always fascinated us was the trends of performance over one's high school career, especially among boys and girls. Discussed among many runners is the fact that men tend to improve their times much more significantly then girls over their high school career. In fact, some even think that it should be expected that a girl decreases in level of performance over their high school career. Considering that tullyrunners.com hosts all of this data on high school track performances, we decided to analyze the data to see whether or not this hypothesis was true and what factors influenced this outcome.
<br>
<p> As we discussed what factors determine whether or not some improves in high school, we came across a few articles that will be discussed in our Related Work section to help us isolate what variables we would be looking for. Combined with our intuition from participating in the sport ourselves, we decided that we observe the following variables along with sex to determine the expected level of change in performance among high school runners: starting performance, school size, level of competition (section and places in races), coaching (whether or not a runner goes to a top school), as well as binary data on whether they improved between each year.
<br>
<p> Our hypothesis is that sex plays a large role in whether or not one improves in high school, but so does coaching and size of school. We believe that there must be a reason that schools like Fayeteville Manlius and Saratoga in New York are dominant in State-level, and National-level competition, so coaching should be a large factor as well as school size (practicing with large amounts of runners). We also believe that we must take a careful look at starting level, as one is much more likely to improve if they start very slow, versus if they start at a high speed rating.

## Related Work
<p> The motivation to explore this dataset originated from one of our team members seeing an article on a track-related site and wanting to see the validity of it. The article entitled: __Running at the Top: How Much Faster Will You Be This Fall?__ (<http://ny.milesplit.com/articles/130667-rtn-001-running-at-the-top-how-much-faster-will-you-be-this-fall#.VIeleYvmmL1>) looked at top New York State runners and how much they improved from year to year. The article found that among top runners, "The elite boys average almost a 3 point gain on average, but the girls slip back at a rate of more than a point."
<br>
<p>The article speculated that the factors that are behind improvement from year to year include "age, the size of their school, stage of physical development, and gender ID." Specifically, the article used narratives to explain why top boys saw average increases in their performance while girls did not. For guys, he speculated that they "have small to moderate gains each year from their sophomore to senior seasons as they develop into bigger and stronger and wiser athletes." On the otherhand, he believes that girls have their own set of issues. He says that "some in fact are at their fastest when they are middle schoolers" and thus have to "battle the psychological strains of seeing fast new younger runners passing by them on their team and at the meets."
<br>
<p>In addition to this specific article, many discussion forums on track on sites such as http://www.tracktalk.net/ and http://www.letsrun.com/forum/forum.php?board=1&category=high_school seem to confirm this notion that a large factor that influences your improvement to year to year is sex, as well as speculating other major reasons that we wanted to explore. Some of these include whether you go to a large school with a large amount of intracompetition, whether you are in a competitive section, and whether you go to a top running program that consistently qualifies for the championship portion of the season. In the discussions, the top school question was speculated the most.

## Initial Questions

1. Given a runner's information (year, initial speed rating, gender, school, etc), how well can we predict how much or little his or her performance will change?
<br>
2. Given information on a runner's performance (especially patterns related to change over years in performance), how well can we predict if the runner is a boy or a girl?
<br>
3. Do elements outside of one's performance effect how one's performance changes in high school such as school size, going to a top school, area in which one competes in?
<br>
<p>These questions remained fairly constant throughout the course of our analysis, though we began to reconsider how to classify certain independent variables and also which new information we could add to see what affected performance change. For example, we quickly found out that boys speed ratings are considerably higher than girls, and thus including initial speed rating would much too easily predict if a runner is a boy or girl, while we could refine relative measures to still answer this question. Another modification was grouping improvement into classes and making them relative. Initially, we were just looking at whether or not a runner improved over his career but found that this did not fully capture our data. For example, some people fell a mere one or two points (a matter of seconds) over their career and they were categorized with runner that fell 80 points. Next, regarding this same question, we realized that performane increases were relative to ability as improving from a high rating is much more difficult than from a low one. In fact, speed rating change the orientation of level of performance. For example a low time in cross country such as 16 minutes is more desirable than 17 minutes, but the correlation is reversed for their corresponding speed rating as 180 (roughly equivalent to 16 minutes) is a better speed rating than 160 (roughly equivalent to 17 minutes). Thus, when we normalized our classifications to incorporate that is more difficult to shed time from less amount of time than more amount of time, we were able to output more accurate results.

###Find Summary Statistics for Each Runner and Filter Results
We want to find runners who have ran enough years that we can analyze performances consecutively for them. Since freshmen typically run in the freshmen races that span only 1.5 miles instead of the standard 5K, there is only a few runners who have four years of consecutive 5K data. Instead, we filter for runners who have three years of data.

In addition, we are only interested in a summary of each of their years. We originally were going to find the median performance for each runner, but upon further consideration, we decided that max of that year would be a more appropriate measure. This is because the majority of a runner's season is running inconsequential meets where top runners can win without running their personal best speed ratings and worse runners also do not run to their best ability because of the slowed pace at the front of the pack. Also, one's ability in track is measured by one's PR/PB (personal record or personal best) and thus we decided max would be a more appropriate measure.

```python
######Girls
median_speed = girls_2013.groupby(['Name','Grade']).max()

#filter for only girls that have run 3 years
only_some = median_speed.groupby(level='Name').filter(lambda x: len(x) > 2)
only_some.head()


########Boys
median_bspeed = boys_2013.groupby(['Name','Grade']).max()

#filter for only girls that have run 3 years
bonly_some = median_bspeed.groupby(level='Name').filter(lambda x: len(x) > 2)
bonly_some.head()
```

#### Normalized Improvment Classes
As mentioned, we are normalizing the improvement classes to depict the degree of performance increase to a standard closer to decreasing time rather than increasing speed rating. These classes are normalized relative to the top observable speed rating (200).

```python
normalized = []
for index, row in master.iterrows():
    n = row['diff1_3'] / (200 - row['1_Rating'])
    
    if n <= -0.45:
        normalized.append(0)
    elif n <= -0.15:
        normalized.append(1)
    elif n <= 0.15:
        normalized.append(2)
    elif n <= 0.45:
        normalized.append(3)
    else:
        normalized.append(4)

master['Normalized Improvement'] = normalized
    
plt.hist(normalized)
plt.show()

master.head()
```

#### Indicate Speed Categories for Starting and Ending Years

```python
category_1 = []
category_3 = []

from math import floor

for index, row in master.iterrows():
    first = row['1_Rating']
    third = row['3_Rating']
    
    if first < -20:
        category_1.append(-2)
    elif first < 0:
        category_1.append(-1)
    else:
        category_1.append(int(floor(first)) / 20)
    
    if third < -20:
        category_3.append(-2)
    elif third < 0:
        category_3.append(-1)
    else:
        category_3.append(int(floor(third)) / 20)

master['category_1'] = category_1
master['category_3'] = category_3

master.head()
```

### KNN For Master Frame

```python
x_train, x_test, y_train, y_test = sklearn.cross_validation.train_test_split(X, Y, test_size = 0.33)
print (x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier

def grid_search(k):
    parameters = {"n_neighbors": np.arange(2, k + 1)}
    clf = KNeighborsClassifier()
    gs = GridSearchCV(clf, param_grid=parameters, cv=10)
    gs.fit(x_train, y_train)
    print gs.best_params_, gs.best_score_
    y_true, y_pred = y_test, gs.predict(x_test)
    print(classification_report(y_true, y_pred))
    return gs.grid_scores_, gs.best_params_

grid_scores_10, params_dict = grid_search(10)
```

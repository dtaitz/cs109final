# Trends in Performance over High School Careers in NYS XC

## Background and Motivation

<p>In high school, 2 of our 3 project members participated in track and cross country and the third was an avid fan of sports analysis. As running is a largely mental sport, it can be very consuming for the individual and thus many athletes try to find trends to help them improve as a runner or predict results from year to year. Luckily, since track and cross country are timed events, data is easily observed and also maintained by a host of different websites. In New York State, where David participated in high school cross country, [Tully Runners](http://www.tullyrunners.com), [NY Milesplit](http://www.ny.milesplit.com), and [Dyestat](http://www.dyestat.com) hold a large database of results. In the case of [Tully Runners](http://www.tullyrunners.com), the founder of the website also conveniently calculates "speed ratings" on cross country results to normalize for the various variables (course difficulty, level of competition, weather, etc) that can affect one's time on a cross country 5K course. His modeling of cross country has been widely accepted in the track and field community and thus a reliable measure on which to run analysis.
<br>
<p> One particular trend that always fascinated us was the trends of performance over one's high school career, especially among boys and girls. Discussed among many runners is the fact that men tend to improve their times much more significantly then girls over their high school career. In fact, some even think that it should be expected that a girl decreases in level of performance over their high school career. Considering that [Tully Runners](http://www.tullyrunners.com) hosts all of this data on high school track performances, we decided to analyze the data to see whether or not this hypothesis was true and what factors influenced this outcome.
<br>
<p> As we discussed what factors determine whether or not some improves in high school, we came across a few articles that will be discussed in our Related Work section to help us isolate what variables we would be looking for. Combined with our intuition from participating in the sport ourselves, we decided that we observe the following variables along with sex to determine the expected level of change in performance among high school runners: starting performance, school size, level of competition (section and places in races), coaching (whether or not a runner goes to a top school), as well as binary data on whether they improved between each year.
<br>
<p> Our hypothesis is that sex plays a large role in whether or not one improves in high school, but so does coaching and size of school. We believe that there must be a reason that schools like Fayeteville Manlius and Saratoga in New York are dominant in State-level, and National-level competition, so coaching should be a large factor as well as school size (practicing with large amounts of runners). We also believe that we must take a careful look at starting level, as one is much more likely to improve if they start very slow, versus if they start at a high speed rating.

## Related Work
<p> The motivation to explore this dataset originated from one of our team members seeing an article on a track-related site and wanting to see the validity of it. The article entitled: __Running at the Top: How Much Faster Will You Be This Fall?__ (<http://ny.milesplit.com/articles/130667-rtn-001-running-at-the-top-how-much-faster-will-you-be-this-fall#.VIeleYvmmL1>) looked at top New York State runners and how much they improved from year to year. The article found that among top runners, "The elite boys average almost a 3 point gain on average, but the girls slip back at a rate of more than a point."
<br>
<p>The article speculated that the factors that are behind improvement from year to year include "age, the size of their school, stage of physical development, and gender ID." Specifically, the article used narratives to explain why top boys saw average increases in their performance while girls did not. For guys, he speculated that they "have small to moderate gains each year from their sophomore to senior seasons as they develop into bigger and stronger and wiser athletes." On the other hand, he believes that girls have their own set of issues. He says that "some in fact are at their fastest when they are middle schoolers" and thus have to "battle the psychological strains of seeing fast new younger runners passing by them on their team and at the meets."
<br>
<p>In addition to this specific article, many discussion forums on track on sites such as http://www.tracktalk.net/ and http://www.letsrun.com/forum/forum.php?board=1&category=high_school seem to confirm this notion that a large factor that influences your improvement to year to year is sex, as well as speculating other major reasons that we wanted to explore. Some of these include whether you go to a large school with a large amount of intracompetition, whether you are in a competitive section, and whether you go to a top running program that consistently qualifies for the championship portion of the season. In the discussions, the top school question was speculated the most.

## Initial Questions

1. Given a runner's information (year, initial speed rating, gender, school, etc), how well can we predict how much or little his or her performance will change?
<br>
2. Given information on a runner's performance (especially patterns related to change over years in performance), how well can we predict if the runner is a boy or a girl?
<br>
3. Do elements outside of one's performance effect how one's performance changes in high school such as school size, going to a top school, area in which one competes in?
<br>
<p>These questions remained fairly constant throughout the course of our analysis, though we began to reconsider how to classify certain independent variables and also which new information we could add to see what affected performance change. For example, we quickly found out that boys' speed ratings are considerably higher than those of girls, and thus including initial speed rating would much too easily predict if a runner is a boy or girl, while we could refine relative measures to still answer this question. Another modification was grouping improvement into classes and making them relative. Initially, we were just looking at whether or not a runner improved over his career but found that this did not fully capture our data. For example, some people fell a mere one or two points (a matter of seconds) over their career and they were categorized with runner that fell 80 points. Next, regarding this same question, we realized that performane increases were relative to ability as improving from a high rating is much more difficult than from a low one. In fact, speed rating change the orientation of level of performance. For example a low time in cross country such as 16 minutes is more desirable than 17 minutes, but the correlation is reversed for their corresponding speed rating as 180 (roughly equivalent to 16 minutes) is a better speed rating than 160 (roughly equivalent to 17 minutes). Thus, when we normalized our classifications to incorporate that is more difficult to shed time from less amount of time than more amount of time, we were able to output more accurate results.

###Find Summary Statistics for Each Runner and Filter Results
We want to find runners who have ran enough years that we can analyze performances consecutively for them. Since freshmen typically run in the freshmen races that span only 1.5 miles instead of the standard 5K, there is only a few runners who have four years of consecutive 5K data. Instead, we filter for runners who have three years of data.

In addition, we are only interested in a summary of each of their years. We originally were going to find the median performance for each runner, but upon further consideration, we decided that max of that year would be a more appropriate measure. This is because the majority of a runner's season is running inconsequential meets where top runners can win without running their personal best speed ratings and worse runners also do not run to their best ability because of the slowed pace at the front of the pack. Also, one's ability in track is measured by one's PR/PB (personal record or personal best) and thus we decided max would be a more appropriate measure.

```python
######Girls
median_speed = girls_2013.groupby(['Name','Grade']).max()

#filter for only girls that have run 3 years
only_some = median_speed.groupby(level='Name').filter(lambda x: len(x) > 2)
only_some.head()


########Boys
median_bspeed = boys_2013.groupby(['Name','Grade']).max()

#filter for only girls that have run 3 years
bonly_some = median_bspeed.groupby(level='Name').filter(lambda x: len(x) > 2)
bonly_some.head()
```

```python
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier

def grid_search(k):
    parameters = {"n_neighbors": np.arange(2, k + 1)}
    clf = KNeighborsClassifier()
    gs = GridSearchCV(clf, param_grid=parameters, cv=10)
    gs.fit(x_train, y_train)
    print gs.best_params_, gs.best_score_
    y_true, y_pred = y_test, gs.predict(x_test)
    print(classification_report(y_true, y_pred))
    return gs.grid_scores_, gs.best_params_

grid_scores_10, params_dict = grid_search(10)
```

Upon thinking about the idea for the final project, we wanted to work with a dataset that could easily be analyzed using regression and machine learning techniques, as those techniques were the heart of what we had learned this semester in CS109. Therefore, we heavily considered incorporating regression techniques using the OLS portion of the statsmodels library in Python as well as machine learning using the sklearn library in Python, with various different classifiers such as Random Forests, Decision Trees, and k-Nearest Neighbors (kNN). ## Final Analysis

The <b>initial quesitons</b> were precisely the following:
1. Given a runner's information (year, initial speed rating, gender, school, etc), how well can we predict how much or little his or her performance will change? 
<br>
2. Given information on a runner's performance (especially patterns related to change over years in performance), how well can we predict if the runner is a boy or a girl?
<br>
3. Do elements outside of one's performance affect how one's performance changes in high school such as school size, going to a top school, area in which one competes in?
<br>

Based on our analysis above, the <b>answers</b> are the following:
<br>
<p>1. Given a runner's information, such as grade, initial speed rating, gender, whether they are in a top running school, and section, we can predict how much or little his or her performance can change with an accuracy of about 60%, using our random forest classifier above. But, this classifier took many iterations to tune and perfect, as we originally thought about just predicting improvement, which was too broad. Then, we decided to divide improvement into different classes so that those who improved more than 20 points throughout their career were put in a different class than those who just improved one point. This metric was hard to classify as an absolute measure. But, when we normalized the improvement data to account for where the person started, since it is much harder to improve once your speed rating is very high, we were able to get our classifier to predict normalized improvement with a 60% accuracy. While there is certainly randomness in an athlete's performance change over years due to factors that cannot be quanitified (motivation) and other that were not given (height/weight), knowing initial speed rating along with gender, school, and place makes it significantly easier to predict one's performance changes. 
<br>
<p>2. Given information on a runner's performance, such as differences between years and ratings in individual years, we can predict if a runner is a boy or a girl pretty well. This is pretty simple, we realized, because an individual's rating in a certain year will pretty much give away their gender, since we are mostly dealing with top runners. Top boys tend to have much higher speed ratings than top girls do, so there is an almost perfect deterministic connection between ratings and sex. But, when we took out ratings in terms of our independent variables, it was much harder to determine whether an individual was a boy or a girl, because of the granularity needed in improvement data (on average 3 points better for boys and 1 point less for boys) to delineate individuals based on gender. While the article in related work focused on only on top runners, we included a larger set of information to see trends for a wider set of the population. There is a great degree more of variability when you add a variety of starting levels for runners. The runner's normalized improvement certainly was the best predictor of this, though. 
<br>
<p>3. We had initially thought that outside factors, such as school size, going to a top school, and section in which an individual competes would affect one's performance quite considerably, because if an individual goes to a top school, we assumed there to be a very good cross-country coach at that school, so the individual would get better. Also, if an individual competes against more runners in practice, then he or she will improve easier, which also applies to the strength of the section in which the individual competes. But, based on the resulting feature importances of our last classifier, the random forest on the normalized improvement data, we actually determined that school size and going to a top school do not have as much of an effect on whether an individual will improve. Therefore, the environment in which an individual practices does not make much of a difference in terms of their levels of improvement. But, the section in which an individual competes does, which means that the strength of a particular region or area affects whether an individual will improve. This result makes sense, as the section determines which people the specific individual will be competing against every week. If an individual has better competition, then he or she will strive to improve more.

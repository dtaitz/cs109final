```python
##last two rows of three is bogus, last row of all
##use skip_footer
#http://pandas.pydata.org/pandas-docs/stable/generated/pandas.io.excel.read_excel.html
girls_2013_all = pd.io.excel.read_excel('XCgirls2013.xlsx', sheetname=1,skip_footer =2 )
girls_2013_three = pd.io.excel.read_excel('XCgirls2013.xlsx', sheetname=0, skip_footer =1 )
```

```python
##now not bogus
##last two rows of three were bogus, last row of all
girls_2013_all[-2:]
girls_2013_three[-1:]
```

```python
#add column for rounded speed rating
girls_2013_all.rename(columns={'Unnamed: 10':'Rounded_Rating'}, inplace=True)
girls_2013_three.rename(columns={'Unnamed: 9':'Rounded_Rating'}, inplace=True)
girls_2013_three.head()
```

```python
##merge results
girls_2013 = pd.merge(girls_2013_all,girls_2013_three, how = 'outer')
girls_2013
```

```python
#girls_2013_three.head()
median_speed = girls_2013.groupby(['Name','Grade']).median()

##median for grade
grade_median= median_speed.groupby(level='Grade', axis = 0).median()
grade_median

#filter for only girls that have run 3 years
#number of people who have ran 4 years is very low
only_some = median_speed .groupby(level='Name').filter(lambda x: len(x) > 2)
only_some.head()
```

```python
##get only the results over the years of seniors
twelve_all = girls_2013_all[girls_2013_all.Grade == 12]
twelve_all_names = set(twelve_all.Name)
senior_mask = girls_2013_all.Name.isin(twelve_all_names)
senior_all = girls_2013_all[senior_mask]
senior_all.head()
```

```python
median_speed_senior = senior_all.groupby(['Name','Grade']).median()
median_speed_senior.head()
```

```python
each_year = only_some.reset_index()
just_ratings = only_some['Rounded_Rating']
rating_school_year= just_ratings.unstack()
rating_school_year.head()
```

```python
no_index = rating_school_year.reset_index()
names_dict = {}

for index,row in rating_school_year.iterrows():
    results_dict = {}
    count = 1
    #print index
    for x in np.arange(7,13,1):
        if pd.notnull(row[x]):
            results_dict[count] = row[x]
            count += 1
    #print results_dict
    names_dict[index] = results_dict

new_df = pd.DataFrame(names_dict).T
only_three = new_df[[1,2,3]]
only_three
list_results = [[row[1],row[2],row[3]] for index,row in only_three.iterrows()]
only_three['list_results'] = list_results
only_three
```
